```python
import geopandas as gp
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
DIR = "./runs/"

runs = []
for f in os.listdir(DIR):
    runs.append(gp.read_file(DIR + f))
```

```python
cmap = plt.get_cmap('Wistia')
colors = [cmap(i) for i in np.linspace(0, 1, len(runs))]

base = runs[0].plot(color=colors[0], figsize=(10,10))

for i, run in enumerate(runs[1:]):
    run = run.to_crs(runs[0].crs)
    plot = run.plot(ax=base, color=colors[i+1])
plot.set_facecolor('black')
```
