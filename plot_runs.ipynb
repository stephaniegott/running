```python
import geopandas as gp
import os
DIR = "./april/"

runs = []
for f in os.listdir(DIR):
    runs.append(gp.read_file(DIR + f))

r1 = runs[0]
r2 = runs[1]
```

```python
#r2.plot(marker='*', color='green', markersize=5);
r2 = r2.to_crs(r1.crs)

base = r1.plot(color='blue')
r2.plot(ax=base, color='red');
```
